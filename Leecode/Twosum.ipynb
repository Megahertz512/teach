## Two sums
Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

 

Example 1:

Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].
Example 2:

Input: nums = [3,2,4], target = 6
Output: [1,2]
Example 3:

Input: nums = [3,3], target = 6
Output: [0,1]
 

Constraints:

2 <= nums.length <= 104
-109 <= nums[i] <= 109
-109 <= target <= 109
Only one valid answer exists.
 

Follow-up: Can you come up with an algorithm that is less than O(n2) time complexity?

```python
class Solution(object):
    def twoSum(self, nums, target):
        enumerated_list = list(enumerate(nums))
        sorted_list = sorted(enumerated_list, key=lambda x: x[1])
        nums = []
        sorted_indices = []
        for i , j in sorted_list:
            nums.append(j)
            sorted_indices.append(i)

        for i in range(len(nums)):
            a = nums[:]
            t1 = a.pop(i)
            if self.binary_search(a , target - t1) != -1:
                return sorted([sorted_indices[i] , sorted_indices[self.binary_search(a , target - t1) + 1]])


    @staticmethod
    def binary_search(lis , x ,L=0 , M = None):

        if M == None:
            M = len(lis) - 1
        
        if M >= L:    
            m = (L + M) // 2
            if lis[m] == x:
                return m
            elif lis[m] > x:
                return Solution.binary_search(lis , x , L , m - 1)
            else:
                return Solution.binary_search(lis , x , m + 1, M)
        else:
            return -1
```

```python
class Solution(object):
    def twoSum(self, numbers, target):
        """
        :type numbers: List[int]
        :type target: int
        :rtype: List[int]
        """
        # sorted in non decreasing order
        # 2 number added to specific target
        # indexes + 1
        visited = {}
        n = len(numbers)

        for i in range(n):
            num = numbers[i]
            if target - num in visited:
                indexI, indexJ = visited[target - num][-1] + 1, i + 1
                return [indexI, indexJ]
            if num not in visited:
                visited[num] = [i]
            else:
                visited[num] += [i]

        return []
```
